```python
df = pd.read_xml('https://raw.githubusercontent.com/Pankaj-khator/InventoryAnalysis/main/Files/walmart_sitemap-category-en0.xml')
```

```python
df.columns = ['URL_FULL','DATE']
```

```python
df[['URL','cat_id']] = df['URL_FULL'].str.rsplit('/',  n=1, expand = True) 
```

```python
df.loc[2,'URL_FULL']
```

df['URL'].str.split('/en/', expand=True)

```python
df[['Website', 'cat']] = df['URL'].str.split('/en/', expand=True)
```

```python
df['cat'] = df['cat'].str.replace('browse/',"")
```

```python
df['cat'].str.split('/', expand = True)
```

```python
df[['cat_lvl_1','cat_lvl_2','cat_lvl_3','cat_lvl_4','cat_lvl_5','cat_lvl_6','cat_lvl_7']]=df['cat'].str.split('/', expand = True)
```

```python
del df['cat']
df
```

```python
frozen_food.drop(columns=['index'], inplace=True)
```
