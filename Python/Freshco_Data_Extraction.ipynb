```python
All_Data = []
storeID = 9636 # Niagara Falls
pagenum =  0 # Page Number

while True:
    payload['requests'][0]['page'] = pagenum
    r = requests.post(url = base_url, headers= headers, json = payload)
    data = r.json()

    hits = (
        data.get('results',[0])[0].get("hits",[])
        )

    if not hits:
        break

    for hit in hits:
        All_Data.append(
            {
                "name" : hit.get('name'),
                "brand" : hit.get('brand'),
                "price": hit.get('price'),
                "articleNumber" : hit.get('articleNumber'),
                "itemAmountUnit" : hit.get('itemAmountUnit'),
                "itemAmountValue" : hit.get('itemAmountValue')
            }
        )
    hitperpage = data.get('results',[0])[0].get('hitsPerPage') 
    nb_of_Hits = data.get('results',[0])[0].get('nbHits') 
    nb_of_Pages = data.get('results',[0])[0].get('nbPages')

    if pagenum == nb_of_Pages :
        break

    pagenum += 1





    
    
```
